### Aberrant (rogue) Values
Now we will traet the aberrant (rogue) data.

We will use the upper and lower quartile method.

```python
Data_train
```

We will display the following graphs so we can visualize the aberrant values.

```python
Data_train.plot(kind='box', subplots=True, layout=(3,3), sharex=False, sharey=False, figsize=(15,10))
plt.show()
```

We notice that there is multiple anormal values, we will create a function that femoves the outliers values.

```python
def removeOutLiers(df,cols):
    Q1 = df[cols].quantile(0.25)
    Q3 = df[cols].quantile(0.75)
    IQR = Q3 - Q1

    df1 = df[~((df[cols] < (Q1 - 1.5 * IQR)) |(df[cols] > (Q3 + 1.5 * IQR))).any(axis=1)]
    return df1
```

```python
AllCountinuousColumns = ['age', 'fnlwgt','educational-num','capital-gain','capital-loss','hours-per-week']
```

```python
df1_train = removeOutLiers(Data_train,AllCountinuousColumns)
```

### Creation and optimization of the model

```python
from sklearn.model_selection import train_test_split
```

First we will divide our data to target and feature data

```python
X = Dfclean_train.drop(['income'] , axis=1)
y = Dfclean_train['income']
```

```python
print("all features : ",X.shape)
print("target variable",y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=22)
```

```python
from sklearn.tree import DecisionTreeClassifier 
from sklearn.neighbors import KNeighborsClassifier
from sklearn import model_selection
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
seed = 19
performance = 'accuracy'
models = []
models.append(('KNN', KNeighborsClassifier())) 
models.append(('CART', DecisionTreeClassifier()))
results = []
names = []
for name, model in models:
    kfold = model_selection.KFold(n_splits=6, random_state=seed, shuffle = True)
    cv_results = model_selection.cross_val_score(model, X_train, y_train, cv=kfold, scoring=performance) 
    results.append(cv_results)
    names.append(name)
    msg = "%s: %f (%f)" % (name, cv_results.mean(), cv_results.std())
    print(msg)
```

We will apply the KNN model on the validation set.

```python
knn = KNeighborsClassifier()
# la méthode fit() réalise généralement la phase d’entrainementàcréation de modèle. Mais, puisque # l’algorithme des K-NN est un lazy learner (pas d’apprentissageàpas de modèle), alors le modèle créé # par fit() est l’ensemble d’apprentissage qui est stocké d’une manière qui accélère le calcul des
# distances (trouver les plus proches voisins).
knn.fit(X_train, y_train)
predictions = knn.predict(X_test) 
print(accuracy_score(y_test, predictions)) 
print(confusion_matrix(y_test, predictions)) 
print(classification_report(y_test, predictions))
```

As the third model we will choose support vector machine model since it is a simple and easy model to use, interpret and more rebuste in the sense that there is little risk of overlearning

```python
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
```

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto'))
clf.fit(X_train, y_train)
```

```python
predictions = clf.predict(X_test) 
print(accuracy_score(y_test, predictions))
```

If we compare the three classification models, we can say that the support vector machine model id the most performant model for our case
